## 7. В наборе auto1 есть безымянные пустые столбцы, исправьте это

```python
auto1 = auto1.drop(auto1.columns[-5:], axis=1)
auto1.head(10)
```

## 8. Объедините auto1 и auto2 в один датафрейм auto

```python
auto = pd.concat([auto1, auto2], ignore_index=True)
auto.iloc[195 : 206]
```

## 9. В датафрейме auto отсутствует колонка owners (владельцы). Создайте серию случайных чисел от 10 000 до 60 000 и добавьте столбец owners в датафрейм auto


```python
auto['owners'] = np.random.randint(10000, 60001, size=len(auto))
auto.head(10)
```

## 10. Каков временной интервал для рассматриваемых моделей?

```python
auto.describe()
```

Временной интервал: 1970-1982 гг.

## 11. Как в течение этого времени изменялось значение расхода топлива?

```python
mpg_mean = auto.groupby('model', as_index=False)['mpg'].mean()

plt.bar(mpg_mean['model'], mpg_mean['mpg'])

plt.xlabel('Год выпуска')
plt.ylabel('Средний расход топлива')
plt.title('Изменение расхода топлива по годам выпуска авто')
```

## 12. Какие модели предпочитали покупатели (TOP-5)?

```python
auto.nlargest(5, 'owners')
```

## 13. Связан ли расход топлива с количеством цилиндров?

```python
mpg_mean = auto.groupby('cylinders', as_index=False)['mpg'].mean()

plt.bar(mpg_mean['cylinders'], mpg_mean['mpg'])

plt.xlabel('Кол-во цилиндров')
plt.ylabel('Средний расход топлива')
plt.title('Изменение расхода топлива в зависимости от кол-ва цилиндров')
```

Ответ: Да, связан.

## 14. Докажите, что расход топлива зависит (или не зависит) от веса автомобиля.

```python
mpg_mean = auto.groupby('weight', as_index=False)['mpg'].mean()

plt.bar(mpg_mean['weight'], mpg_mean['mpg'])

plt.xlabel('Вес автомобиля')
plt.ylabel('Средний расход топлива')
plt.title('Изменение расхода топлива в зависимости от веса автомобиля')
```

По графику видно, что расход зависит от веса автомобиля.
